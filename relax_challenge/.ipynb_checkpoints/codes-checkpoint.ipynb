1] A user table ( "takehome_users" ) with data on 12,000 users who signed up for the
product in the last two years. This table includes:
● name: the user's name
● object_id: the user's id
● email: email address
● creation_source: how their account was created. This takes on one
of 5 values:
○ PERSONAL_PROJECTS: invited to join another user's
personal workspace
○ GUEST_INVITE: invited to an organization as a guest
(limited permissions)
○ ORG_INVITE: invited to an organization (as a full member)
○ SIGNUP: signed up via the website
○ SIGNUP_GOOGLE_AUTH: signed up using Google
Authentication (using a Google email account for their login
id)
● creation_time: when they created their account
● last_session_creation_time: unix timestamp of last login
● opted_in_to_mailing_list: whether they have opted into receiving
marketing emails
● enabled_for_marketing_drip: whether they are on the regular
marketing email drip
● org_id: the organization (group of users) they belong to
● invited_by_user_id: which user invited them to join (if applicable).

2] A usage summary table ( "takehome_user_engagement" ) that has a row for each day
that a user logged into the product.

Defining an "adopted user" as a user who has logged into the product on three separate days in at least one sevenday period , identify which factors predict future user adoption .

```python
import numpy as np

# features: list of variables names between which correlation is calculated
def ANOVA(groupby, col, df):
    data=df[[groupby, col]].dropna()
    n=len(data[groupby].unique())
    N=data.shape[0]
    SSwithin = sum(data.groupby(groupby).var()[col]*data[groupby].value_counts())
    fw=N - n
    SSbetween = sum(((data.groupby(groupby).mean()[col]-data[col].mean())**2)*data[groupby].value_counts())
    fb = n - 1
    MSbetween = SSbetween/fb
    MSwithin = SSwithin/fw
    F = MSbetween/MSwithin
    p = scipy.stats.f.sf(F, fb, fw)
    return (p)

def Pearsonr(col1,col2,df):
    data=df[[col1,col2]].dropna()
    n=data.shape[0]
    sum_cov=((data[col1]-data[col1].mean())*(data[col2]-data[col2].mean())).sum()
    std1=(((data[col1]-data[col1].mean())**2).sum()/(n-1))**0.5
    std2=(((data[col2]-data[col2].mean())**2).sum()/(n-1))**0.5
    r=(1/(n-1))*sum_cov/(std1*std2)
    t=r*(n-2)**(0.5)/(1-r**2)**(0.5)
    p_value=scipy.stats.t.sf(t, n-1)
    return (r,p_value)

def Correlation(features, df, printout=False):
    category=["category", "bool", 'object']
    continuous=['int64', 'float64', 'int32', 'float32']
    if features[0]==features[1]:
        print ("same features!")
    else:
        if df[features[0]].dtype.name in category and df[features[1]].dtype.name in category:
            table=pd.crosstab(df[features[0]], df[features[1]]).values
            table=table[~np.all(table == 0, axis=1)]
            p_value=scipy.stats.chi2_contingency(table)[1]
            if printout:
                print ("\n-----------------------------------------\n")
                total_groupby=df_com.groupby(features[0]).agg({features[1]:"value_counts"})
                pcts_groupby=total_groupby.groupby(level=0).apply(lambda x:
                                                 100 * x / float(x.sum())).unstack()
                print (pcts_groupby.sort_values(ascending=False))
                print ("chi-square test p value is", p_value)
                print ("\n-----------------------------------------\n")
            else:
                return (p_value)
        elif df[features[0]].dtype.name in category and df[features[1]].dtype.name in continuous:
            p_value=ANOVA(features[0], features[1], df)
            if printout:
                sns.barplot(x=features[0], y=features[1], data=df, label='p value of ANOVA is {}'.format(round(p_value,4)))
                plt.legend()
                plt.show()
            else:
                return (p_value)
        elif df[features[0]].dtype.name in continuous and df[features[1]].dtype.name in category:
            p_value=ANOVA(features[1], features[0], df)
            if printout:
                sns.barplot(x=features[1], y=features[0], data=df, label='p value of ANOVA is {}'.format(round(p_value,4)))
                plt.legend()
                plt.show()
            else:
                return (p_value)
        elif df[features[0]].dtype.name in continuous and df[features[1]].dtype.name in continuous:
            
            r,p_value=Pearsonr(features[0], features[1],df)
            if printout:
                sns.regplot(features[0], features[1],df, label='pearson correlation={}, p value={}'.format(r, p_value))
                plt.xlabel(features[0])
                plt.ylabel(features[1])
                plt.legend()
                plt.show()
            else:
                return (p_value)

def Correlation_matrix(features, df):
    from itertools import combinations
    correlation_matrix=np.empty([len(features), len(features)])
    corr_df=pd.DataFrame(correlation_matrix)
    corr_df.index=features
    corr_df.columns=features
    for tup in combinations(features, 2):
        p_value=Correlation([tup[0],tup[1]], df)
        corr_df.loc[tup[0]][tup[1]]=corr_df.loc[tup[1]][tup[0]]=round(p_value, 4)

    return (corr_df)
```

### Classification- test-train split, logistic regression and model valication using metrics

No severe multicollinearity is observed based on the criteria vif<2.5. Dummy variables have inf vif because they are created from the same categorical variable. Then I use logistic regression model to understand which is an important predictor for adopted user.

It turns out the data has unbalanced distribution. Therefore I used stratified sampling to randomly draw training and test data from adopted users and non-adopted users

```python
def Stratified_sample(outcome, df, testsize):
    X_trains=list()
    y_trains=list()
    X_tests=list()
    y_tests=list()
    
    for v in df[outcome].unique():
        X=df[df[outcome]==v].drop([outcome], axis=1).values
        y=df[df[outcome]==v][outcome].astype(int).values
           
        X_trainset, X_testset, y_trainset, y_testset = train_test_split(X, y, test_size=testsize, random_state=42)
        X_trains.append(X_trainset)
        y_trains.append(y_trainset)
        X_tests.append(X_testset)
        y_tests.append(y_testset)
        
    X_train=np.row_stack(X_trains)
    y_train=np.concatenate(y_trains)

    X_test=np.row_stack(X_tests)
    y_test=np.concatenate(y_tests)    
    
    return (X_train, X_test, y_train, y_test)
```

```python
def Sampling_by_weight(X_train,y_train,sampleweight):
    indices=list()
    for v in sampleweight.keys():
        idx_list=np.argwhere(y_train==v)[:,0]
        print (idx_list)
        indices.extend(np.random.choice(idx_list, int(len(idx_list)*sampleweight[v])))
    X=X_train[indices][:]
    y=y_train[indices][:]
    
    return (X,y)
```

```python
from sklearn.model_selection import train_test_split
X_train, X_test, y_train, y_test=Stratified_sample('adopted', df_dummy, testsize=0.25)
```

```python
#rescale the columns (except dummy variables) to have mean at zero and variance around 1
def Rescale(X):
    data=pd.DataFrame(X)
    for c in data.columns:
        if set(data[c]) !=set([0,1]):
            data[c]=(data[c]-data[c].mean())/data[c].std()
    return (data.values)
```

```python
X_train=Rescale(X_train)
X_test=Rescale(X_test)
```

```python
from sklearn.model_selection import StratifiedKFold
from sklearn.linear_model import LogisticRegression
from sklearn.model_selection import GridSearchCV
from sklearn.metrics import precision_recall_curve, roc_curve, auc, confusion_matrix, precision_score, recall_score, f1_score

def Gridsearchcv_report(data, classifier, param_grid, scores, refit_score):
    """grid search for a classifier that returns the searched results"""
    X_train, X_test, y_train, y_test=data
    stk=StratifiedKFold(n_splits=10)
    GCV= GridSearchCV(classifier, param_grid, scoring=scores, refit=refit_score,cv=stk, return_train_score=True, n_jobs=-1)
    GCV.fit(X_train, y_train)
    
    
    
    print ("\n----------------------------------------------\n")
    print ("When optimized for {} on the training data:".format(refit_score))
    print ("tuned parameter are {}".format(GCV.best_params_))
    print ("The final {} is {}".format(refit_score, GCV.best_score_))
    
    print('\nConfusion matrix of training data:')
    y_pred = GCV.predict(X_train)
    y_score=GCV.decision_function(X_train)
    print(pd.DataFrame(confusion_matrix(y_train, y_pred),
                 columns=['predicted neg', 'predicted pos'], index=['true neg', 'true pos']))
    f1=f1_score(y_train, y_pred)
    precision=precision_score(y_train, y_pred)
    recall=recall_score(y_train, y_pred)
    fpr, tpr, _ = roc_curve(y_train, y_score)
    AUC=auc(fpr, tpr)
    print ('\nmetrics on the train data are:\nprecision={}, recall={}, f1_meansure={}, AUC={}'.format(precision, recall, f1, AUC))
    
    y_pred = GCV.predict(X_test)
    y_score=GCV.decision_function(X_test)
    print('\nConfusion matrix of test data:')
    print(pd.DataFrame(confusion_matrix(y_test, y_pred),
                 columns=['predicted neg', 'predicted pos'], index=['true neg', 'true pos']))
    f1=f1_score(y_test, y_pred)
    precision=precision_score(y_test, y_pred)
    recall=recall_score(y_test, y_pred)
    fpr, tpr, _ = roc_curve(y_test, y_score)
    AUC=auc(fpr, tpr)
    
    print ('\nmetrics on the test data are:\nprecision={}, recall={}, f1_meansure={}, AUC={}'.format(precision, recall, f1, AUC))
    print ("\n----------------------------------------------\n")
    
    return (GCV)
```

```python
LR=LogisticRegression(class_weight='balanced')
#param_grid = [{"C": np.linspace(0.001, 1000, 30),
              # "intercept_scaling": [1, 3, 5, 10, 20, 50, 100, 200]}]
param_grid = [{"C": np.linspace(0.001, 1000, 40)}]
scores = ['roc_auc','precision_macro', 'recall_macro', 'f1_macro']
data=(X_train, X_test, y_train, y_test)
GCVs=list()
for score in scores:
    GCV=Gridsearchcv_report(data, LR, param_grid, scores, refit_score=score)
    GCVs.append(GCV)
```

For all four metrics that this model is optimized on, the model never gives a good precision, this can be a result of the imbalanced classes since only 18% of users are adopted in both training and test set. Then I try undersampling the non adopted users. 

```python
X_train, y_train=Sampling_by_weight(X_train,y_train,sampleweight={0:0.23,1:1})
```

```python
print ('after undersampling {} users are adopted users'.format(sum(y_train)/len(y_train)))
```

Rescaling the resampled training set and rerun the logistic regression

```python
X_train=Rescale(X_train)
X_test=Rescale(X_test)
```

```python
LR=LogisticRegression(class_weight='balanced')
#param_grid = [{"C": np.linspace(0.001, 1000, 30),
              # "intercept_scaling": [1, 3, 5, 10, 20, 50, 100, 200]}]
param_grid = [{"C": np.linspace(0.001, 1000, 40)}]
scores = ['roc_auc','precision_macro', 'recall_macro', 'f1_macro']
data=(X_train, X_test, y_train, y_test)
GCVs=list()
for score in scores:
    GCV=Gridsearchcv_report(data, LR, param_grid, scores, refit_score=score)
    GCVs.append(GCV)
```

It shows that after undersampling the non adopted users, the recall and f1_score all increase for the training data, but remains relatively low for the test data. This indicates overfitting which can be a problem of oversampling the adopted users. Following shows the roc curve and precision-recall curve of models optimized on different metrics. These models have almost the same curves. 

```python
fig=plt.figure()
for i in range(4):
    GCV=GCVs[i]
    y_score = GCV.decision_function(X_test)
    fpr, tpr, _ = roc_curve(y_test, y_score)
    plt.plot(fpr, tpr, label='model'+str(i+1))
    plt.xlabel('false positive rate')
    plt.ylabel('true positive rate')
plt.plot([0,1],[0,1], color='r', linestyle='--')
plt.legend()
plt.show()

fig=plt.figure()
for i in range(4):
    GCV=GCVs[i]
    y_pred = GCV.predict(X_test)
    precision, recall, _ = precision_recall_curve(y_test, y_pred, pos_label=1)
    plt.plot(precision, recall, label='model'+str(i+1))
    plt.xlabel('precision')
    plt.ylabel('recall (true positive rate)')
plt.legend()
plt.show()
```

Further techniques could be used to improve the model performance. However, I will stop here to get some insights from the results. 

```python
import seaborn as sns

def Coefficient_plot(coef, names):
    coef_db = pd.DataFrame(np.round_(coef, decimals=3), columns = ["coefficients"])
    coef_db['features']=names     
    # remove the zero coefficients
    coef_db = coef_db[coef_db['coefficients'] != 0]
    # sort the values from high to low
    coef_db = coef_db.sort_values(by = 'coefficients', ascending = False)
    plt.figure(figsize=(20,60))
    ax = sns.barplot(x = 'coefficients', y= coef_db['features'] , data=coef_db)
    ax.set(xlabel='Penalized Regression Coefficients')
    sns.set(font_scale = 5)
    plt.show()
```

# Some observations: 

### The coefficients show how each feature affect the probability that a user will be adopted or not:
- if a user is invited by an adopted user when this user also more likely to become adopted user; 
- Users who use hotmail, gmail and jourrapide are more likely to be adopted users than those who use yahoo, gustr, cuvox, or other email address. 
- Users who created account earlier are more likely to be adopted. 
- Users who are invited by another user from the same organization are more likely to be adopted users;
- Being from organization with id between 280-290, 380-390 and 230-240 increase the probability of being adopted users. By contrast, being from organization with id 0-10, 360-370, 140-150 etc reduces the probabiilty of being adopted users.

### The insights from the results: 
To increase the probability of users being adopted, the company may work in following directions:
- Promting the chance that adopted users invite new users
- Putting more efforts into attracting users through gmail, hotmail and jourrapide
- Understanding what causes users who create accounts later to be less likely adopted
- Identifying the key features of the organizations and supplement the data with these features to enhance model performance and interpretability
